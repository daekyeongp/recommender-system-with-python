## 데이터 전처리

```python
rating_data.drop('timestamp', axis = 1, inplace = True)
rating_data.head()
```

```python
movie_data.drop('genres', axis = 1, inplace = True)
movie_data.head()
```

```python
user_movie_data = pd.merge(rating_data, movie_data, on = 'movieId')
user_movie_data.head()
```

```python
user_movie_data.shape
```

**불필요한 컬럼을 제거**한 뒤 pandas의 **merge**를 이용해서 하나로 병합.

그 후, pivot_table을 생성.
- value에는 **rating값**을, column에는 **movie title 값**을, index는 **user id값**을 사용

```python
user_movie_rating = user_movie_data.pivot_table('rating', index = 'userId', columns='title').fillna(0)
```

```python
user_movie_rating.shape
```

```python
user_movie_rating.head()
```

생성된 사용자-영화 데이터의 pivot table는 위와 같음.

사용자-영화 기준의 데이터를 **영화-사용자** 기준으로 만들어서 **특정 '영화'와 비슷한 영화를 추천**해주는 로직을 구현해야 함.


```python
movie_user_rating = user_movie_rating.values.T
movie_user_rating.shape
```

```python
type(movie_user_rating)
```

여기서 이제 SVD를 사용합니다.   

**SVD(Singular Value Decomposion), 특이값 분해**란 m x n 크기의 데이터 행렬 A를 아래와 같이 분해하는 것을 말합니다.

![15](https://user-images.githubusercontent.com/24634054/73115129-93138c00-3f65-11ea-9a10-80abc59a8494.JPG)


출처 : https://ratsgo.github.io/from%20frequency%20to%20semantics/2017/04/06/pcasvdlsa/

행렬 U와 V에 속한 열벡터는 특이벡터(singular vector)로 불리고, 이 특이벡터들은 서로 직교하는 성질을 가지고 있음.
- 또한, 가운데 시그마 기호로 보이는 것은 이것도 행렬인데, 이 행렬은 대각행렬(diagonal matrix) 성질을 가지고 있음.
- 즉, 대각 성분이 행렬 A의 특이값이고 나머지 성분이 0 이라고 이해하자.

추가로, 사이킷런에서 제공해주는 truncated SVD는 이러한 SVD의 변형임.
- **truncated SVD**는 시그마 행렬의 대각원소(특이값) 가운데 상위 n개만 골라낸 것임. 
- 이렇게 하면 기존 행렬 A의 성질을 100% 원복할 수는 없지만, (그 만큼 데이터 정보를 압축 함) 행렬 A와 거의 근사한 값이 나오게 됨.

실험은 scikit learn의 **TruncatedSVD**를 사용하여 진행 함.

```python
SVD = TruncatedSVD(n_components=12)
matrix = SVD.fit_transform(movie_user_rating)
matrix.shape
```

```python
matrix[0]
```

12개의 component로 차원을 축소했으며, 생성된 데이터를 사용하여 **피어슨 상관계수**를 구해야 함.

```python
corr = np.corrcoef(matrix)
corr.shape
```

```python
corr2 = corr[:200, :200]
corr2.shape
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(corr2)
```

생성된 상관계수를 이용해서 **특정 영화와 상관계수가 높은** 영화를 뽑아서 추천해야 함.

```python
movie_title = user_movie_rating.columns
movie_title_list = list(movie_title)
coffey_hands = movie_title_list.index("Guardians of the Galaxy (2014)")
```

```python
corr_coffey_hands  = corr[coffey_hands]
list(movie_title[(corr_coffey_hands >= 0.9)])[:50]
```

**마블의 가디언즈 오브 갤럭시** 영화를 기준으로 비슷한 영화를 추출한 결과
